```python
import pandas as pd
```

```python
df = pd.read_excel("sells.xlsx")
df.head()
```

### logistical forecast (classification)

```python
# Assuming dataset is your DataFrame containing both features and target variable
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

### Other way to divide DF

x = df[["media_rating", "media_pedidos", "test"]]

y = df[["cliente_comprou"]]

```python
x.head()
```

```python
x = pd.get_dummies(x)
# Converter valores booleanos para 1 ou 0
x = x.astype(int)
x.head()
```

```python
y.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
tree = tree.DecisionTreeClassifier()
```

```python
model =tree.fit(x,y)
model
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(model,
               feature_names=x.columns,
               class_names=["yes", "no"],
               filled=True)
plt.show()
```
